```python
import pandas as pd
import s3fs

# --- MinIO Configuration ---
# Port 9000 is the API. Port 9001 is the Web UI.
MINIO_API_URL = "http://127.0.0.1:9000" 
BUCKET_NAME = "raw-market-data"

# Update these with the credentials you use to log into the MinIO UI
ACCESS_KEY = "admin" 
SECRET_KEY = "password123"

print(f"Connecting to MinIO API at {MINIO_API_URL}...")

try:
    # 1. Build the S3 connection pointing at your local MinIO
    fs = s3fs.S3FileSystem(
        key=ACCESS_KEY,
        secret=SECRET_KEY,
        client_kwargs={'endpoint_url': MINIO_API_URL}
    )

    # 2. Find the Parquet files inside the bucket
    # Spark usually puts files inside directories, so we look for the actual .parquet files
    files = fs.glob(f"{BUCKET_NAME}/**/*.parquet")
    
    if not files:
        print("No .parquet files found in the bucket. Is Spark writing them yet?")
        exit()

    print(f"Found {len(files)} parquet files. Reading the most recent one...")
    
    # 3. Read the first file into Pandas using the S3 filesystem
    latest_file = f"s3://{files[-1]}"
    df = pd.read_parquet(latest_file, filesystem=fs)
    
    print(f"\n--- Successfully loaded {latest_file} ---")
    print(f"Total Rows: {len(df)}")
    
    # 4. The Investigation: Look for the UNKNOWNs
    print("\n--- Raw Data Preview (First 5 Rows) ---")
    print(df.head())
    
    unknowns = df[df.apply(lambda row: row.astype(str).str.contains('UNKNOWN', case=False).any(), axis=1)]
    
    if not unknowns.empty:
        print(f"\n🚨 Found {len(unknowns)} rows containing 'UNKNOWN':")
        print(unknowns[['ticker', 'price', 'volume']].head(20))
    else:
        print("\n✅ No 'UNKNOWN' strings found in this raw file.")

except Exception as e:
    print(f"\nFailed to connect or read from MinIO: {e}")
```
